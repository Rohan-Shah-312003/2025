```python
import cv2
import matplotlib.pyplot as plt # Import the matplotlib library

# Load the image in color
image = cv2.imread("/content/images/Test/Her(42).jpg")

# Convert to LAB color space
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Split channels
l, a, b = cv2.split(lab)

# Apply CLAHE to L-channel
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
l_clahe = clahe.apply(l)

# Merge back channels
lab_clahe = cv2.merge((l_clahe, a, b))

# Convert back to BGR
enhanced_image = cv2.cvtColor(lab_clahe, cv2.COLOR_LAB2BGR)

# Show original vs enhanced image
plt.figure(figsize=(10,5)) # Now plt is defined and can be used
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB))
plt.title("Enhanced Image")
plt.axis("off")

plt.show()
```

```python
def add_distortion(image):
    rows, cols = image.shape[:2]

    # Define random transformation points
    src_pts = np.float32([[0, 0], [cols-1, 0], [0, rows-1], [cols-1, rows-1]])
    dst_pts = np.float32([[0, 10], [cols-1, -10], [10, rows-1], [cols-10, rows-10]])

    # Compute transformation matrix
    matrix = cv2.getPerspectiveTransform(src_pts, dst_pts)

    # Apply perspective warp
    distorted_image = cv2.warpPerspective(image, matrix, (cols, rows))

    return distorted_image

# Apply distortion
distorted = add_distortion(image)

# Show original vs distorted
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(distorted, cmap='gray')
plt.title("Distorted Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import random
import pandas as pd
from tqdm import tqdm

# Dataset Path
dataset_path = "/content/images"

# Get list of images and class labels
class_counts = {}  # Dictionary to store class distribution
image_files = []

# Traverse dataset directory
for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder):  # Check if it's a folder
        files = [os.path.join(class_folder, f) for f in os.listdir(class_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]
        image_files.extend(files)
        class_counts[class_name] = len(files)

# Limit to a few images for visualization
num_images = min(6, len(image_files))
sample_images = random.sample(image_files, num_images)

# Function to add distortion (Perspective Warp)
def add_distortion(image):
    rows, cols = image.shape[:2]

    # Define transformation points
    src_pts = np.float32([[0, 0], [cols-1, 0], [0, rows-1], [cols-1, rows-1]])
    dst_pts = np.float32([[0, 20], [cols-20, -10], [10, rows-10], [cols-10, rows-20]])

    matrix = cv2.getPerspectiveTransform(src_pts, dst_pts)
    distorted_image = cv2.warpPerspective(image, matrix, (cols, rows))

    return distorted_image

# Plot Heatmaps and Distorted Images
plt.figure(figsize=(15,10))

for i, img_path in enumerate(sample_images):
    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Load grayscale
    image = image / 255.0  # Normalize

    distorted = add_distortion(image)  # Apply distortion

    # Plot original heatmap
    plt.subplot(3, num_images, i+1)
    sns.heatmap(image, cmap="magma", cbar=True)
    plt.title("Heatmap")
    plt.axis("off")

    # Plot distorted heatmap
    plt.subplot(3, num_images, i+1+num_images)
    sns.heatmap(distorted, cmap="magma", cbar=True)
    plt.title("Distorted Heatmap")
    plt.axis("off")

plt.suptitle("Heatmaps and Distorted Images of Egyptian Hieroglyphics", fontsize=16)
plt.show()

# Bar plot for dataset class distribution
df = pd.DataFrame(list(class_counts.items()), columns=["Class", "Count"])
plt.figure(figsize=(10,5))
sns.barplot(x="Class", y="Count", data=df, palette="viridis")
plt.xticks(rotation=45)
plt.title("Class Distribution in Egyptian Hieroglyphics Dataset")
plt.xlabel("Hieroglyph Class")
plt.ylabel("Number of Images")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Path to your single image
image_path = "/content/images/Test/100 (43).jpg"  # Replace with your image file path

# Function to perform histogram equalization
def histogram_equalization(image):
    # If image is in grayscale, perform histogram equalization
    if len(image.shape) == 2:  # Grayscale image
        equalized_image = cv2.equalizeHist(image)  # Equalizing the histogram
    elif len(image.shape) == 3:  # Color image (apply on each channel)
        # Convert to LAB color space for better results
        lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)  # Split into L, A, B channels
        l = cv2.equalizeHist(l)  # Equalize only the L (lightness) channel
        lab = cv2.merge([l, a, b])  # Merge back
        equalized_image = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # Convert back to BGR
    else:
        equalized_image = image  # No equalization for unsupported formats

    return equalized_image

# Read the image (you can read either grayscale or color image)
image = cv2.imread(image_path)

# Perform histogram equalization
equalized_image = histogram_equalization(image)

# Display the original and equalized images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

# Equalized Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.title("Equalized Image")
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define dataset and output paths
dataset_path = "/content/images"  # Set your dataset path
output_path = "output_heatmaps"  # New folder for heatmaps

# Create output folder if it doesn't exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Function to apply distortion (Perspective Warp)
def add_distortion(image):
    rows, cols = image.shape[:2]

    # Define transformation points
    src_pts = np.float32([[0, 0], [cols-1, 0], [0, rows-1], [cols-1, rows-1]])
    dst_pts = np.float32([[0, 20], [cols-20, -10], [10, rows-10], [cols-10, rows-20]])

    matrix = cv2.getPerspectiveTransform(src_pts, dst_pts)
    distorted_image = cv2.warpPerspective(image, matrix, (cols, rows))

    return distorted_image

# Process both 'Train' and 'Test' folders (exact case-sensitive names)
for split in ["Train", "Test"]:  # Use "Train" and "Test" for exact folder names
    split_path = os.path.join(dataset_path, split)  # e.g., "dataset/Train" or "dataset/Test"
    output_split_path = os.path.join(output_path, split)  # e.g., "output_heatmaps/Train"

    # Create output split folder if it doesn't exist
    if not os.path.exists(output_split_path):
        os.makedirs(output_split_path)

    # Process each image in the folder
    for img_file in os.listdir(split_path):
        if img_file.endswith(('.png', '.jpg', '.jpeg')):  # Ignore XML files
            img_path = os.path.join(split_path, img_file)

            # Read image in grayscale
            image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            image = image / 255.0  # Normalize pixel values

            # Apply distortion
            distorted = add_distortion(image)

            # Generate heatmap for distorted image
            plt.figure(figsize=(5, 5))
            sns.heatmap(distorted, cmap="magma", cbar=True)
            plt.axis("off")

            # Save heatmap
            save_path = os.path.join(output_split_path, f"heatmap_distorted_{img_file}")
            plt.savefig(save_path, dpi=300, bbox_inches="tight")
            plt.close()

print(f"✅ All distorted heatmaps saved in '{output_path}' folder with 'Train/' and 'Test/' structure.")
```
